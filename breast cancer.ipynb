```python
data = data.drop(data[data['perimeter_se']>13].index)
data.describe(include='all')
```

```python
clean = {'diagnosis':{'M':1,'B':0}}
data1= data.replace(clean)
data1.head()
```

```python
data1.columns
```

```python
from sklearn.model_selection import train_test_split
x= data1.drop('diagnosis',1)
y= data1[['diagnosis']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x1 = x_train[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
x2 = x_train[['radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se']]
x3 = x_train[['radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
x1.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
model1 = DecisionTreeClassifier()
model1.fit(x1,y_train)
model2 = LogisticRegression()
model2.fit(x2,y_train)
model3 = DecisionTreeClassifier()
model3.fit(x3,y_train)
```

```python
datamodel = pd.DataFrame()
datamodel['x1_pred']= model1.predict(x1)
datamodel['x2_pred']=model2.predict(x2)
datamodel['x3_pred']= model3.predict(x3)
#datamodel['y_actual']=y_train.reset_index(drop=True)
datamodel.head()
```

```python
# building new model
#new_x = datamodel.drop('y_actual',1)
new_x = datamodel
#new_y = datamodel[['y_actual']]
new_model = LogisticRegression()
new_model.fit(new_x,y_train)
```

```python
# Now doing the same for testing dataset
x1_test = x_test[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]
x2_test = x_test[['radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se']]
x3_test = x_test[['radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
datamodel_new = pd.DataFrame()
datamodel_new['x1_pred']= model1.predict(x1_test)
datamodel_new['x2_pred']=model2.predict(x2_test)
datamodel_new['x3_pred']= model3.predict(x3_test)
datamodel_new.head()
```

```python
final_y = new_model.predict(datamodel_new)
```

```python
from sklearn.metrics import plot_confusion_matrix, confusion_matrix, ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_test,final_y)
```

```python
ConfusionMatrixDisplay(cm).plot()
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(classification_report(y_test,final_y))
print(accuracy_score(y_test,final_y))
```
